```python
from datetime import date
from dateutil import relativedelta
import pandas as pd
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
from IPython.display import clear_output, display, HTML
```

```python
class OptionsGrant():
    def __init__(self, vesting_date, qty, strike, split=1):
        self.vesting_date = vesting_date
        self.quantity = qty*split
        self.strike = strike/split
        
    def months_vested(self, on_date=date.today()):
        
        r = relativedelta.relativedelta(on_date, self.vesting_date)
        months = r.years*12 + r.months
        months = max(0, months)
        return months
    
    def pct_vested(self, on_date=date.today()):
        return min(1., self.months_vested(on_date)/48.)
        
    def qty_vested(self, on_date=date.today()):
        return self.quantity * self.pct_vested(on_date)
    
    def cost_vested(self, on_date=date.today()):
        return self.qty_vested(on_date)*self.strike
    
    def value_vested(self, price, on_date=date.today()):
        return self.qty_vested(on_date) * price
    
    def net_value(self, price, on_date=date.today()):
        return self.value_vested(price, on_date=on_date) - self.cost_vested(on_date)

def broadcast(gs, method, *args, **kwargs):
    return map(lambda x: getattr(x, method)(*args, **kwargs), gs)    
```

```python
gs = [OptionsGrant(date(2011,4,18), 3000, 1.89, split=2./3),
      ]
```

```python
def make_plots(price):
    pers = 365*6
    ts = pd.DataFrame(index=pd.DateRange(date(2014,1,1), periods = pers))
    ts = ts.reset_index()
    ts['net_value'] = ts['index'].apply(lambda x: sum(broadcast(gs, 'net_value', price, on_date=x)))
    ts['value'] = ts['index'].apply(lambda x: sum(broadcast(gs, 'value_vested', price, on_date=x)))
    ts['cost'] = ts['index'].apply(lambda x: sum(broadcast(gs, 'cost_vested', on_date=x)))
    ts = ts.set_index('index')
    ts.net_value.plot(figsize=(12,8))
    #ts.cost.plot(figsize=(12,8))
    ts.value.plot(figsize=(12,8))
```

```python
interact(make_plots, #years=widgets.IntTextWidget(value=1), 
         price=widgets.FloatSliderWidget(min=0.0, max=100, step=0.1, value=14.0))
```
